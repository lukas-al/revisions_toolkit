```python
import pandas as pd
import numpy as np
import pathlib
import re
```

```python
def detect_time_series_rows(df):
    # Load the Excel file


    # Find the rows where the leftmost column is a date
    assumed_index = df.iloc[:, 0]
    assumed_index = assumed_index.map(lambda x: ' '.join(re.findall(r'[A-Za-z]+|\d+', x)))
    print(assumed_index)
    date_rows = pd.to_datetime(assumed_index, errors='coerce').notna()
    print(date_rows)
    # Find the rows where there is a number in any of the columns to the right of the date column
    number_rows = df.iloc[:, 1:].map(np.isreal).any(axis=1)

    # The start row is the first row where the leftmost column is a date
    start_row = date_rows.idxmax() if date_rows.any() else None

    # The end row is the last row where there is a number in any of the columns to the right of the date column
    end_row = number_rows[::-1].idxmax() if number_rows.any() else None

    return start_row, end_row
```

```python
test_path = pathlib.Path("/Users/lukasalemu/Downloads/income/CAEQ - Public corporations gross operating surplus.xlsx")
test_path2 = pathlib.Path("/Users/lukasalemu/Downloads/dataset5rftrt.xls")
test_df = pd.read_excel(test_path2, sheet_name='BI_triangle')
```

```python
start_row, end_row = detect_time_series_rows(test_df)
start_row, end_row
```

```python
# Read the excel using the start and end rows to limit the data
df = pd.read_excel(test_path, sheet_name='triangle', skiprows=start_row, nrows=end_row-start_row)
```
