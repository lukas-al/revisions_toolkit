```python
# Our example of a page will be the following
# https://www.ons.gov.uk/economy/inflationandpriceindices/datasets/consumerpriceindices/current

# Unfortunately, the ONS only stores the data from 2016 onwards, with previous releases stored on hte national archives
# I'm going to consider that out of scope for now.

# An example of the file patterns we're looking for is:
# 0: https://www.ons.gov.uk/file?uri=/economy/inflationandpriceindices/datasets/consumerpriceindices/current/mm23.xlsx
# 1: https://www.ons.gov.uk/file?uri=/economy/inflationandpriceindices/datasets/consumerpriceindices/current/previous/v103/mm23.xlsx
# 2: https://www.ons.gov.uk/file?uri=/economy/inflationandpriceindices/datasets/consumerpriceindices/current/previous/v102/mm23.xlsx
# all the way to
# N: https://www.ons.gov.uk/file?uri=/economy/inflationandpriceindices/datasets/consumerpriceindices/current/previous/v4/mm23.xlsx

# Write some code to find all the links to these files and download them to the local directory:

import requests
import re
import pandas as pd
# First, we need to get the page
url = "https://www.ons.gov.uk/economy/inflationandpriceindices/datasets/consumerpriceindices/current"
page = requests.get(url)
page.raise_for_status()
identified_file_links =  re.findall(r'"(/file\?uri=[^"]*.csv)', page.text)
```

```python
df_dict = {}

for link in identified_file_links:
    download_link = "https://www.ons.gov.uk" + link
    
    # Download the files as dataframes into a dictionary where the key is the filename and the value is the dataframe

    df_dict[link] = pd.read_csv(download_link)
    
    
```

```python
import yaml

def convert_conda_to_pip(conda_env_file, pip_requirements_file):
    with open(conda_env_file, 'r') as f:
        env = yaml.safe_load(f)

    with open(pip_requirements_file, 'w') as f:
        for dep in env['dependencies']:
            # Skip any dependency that is a dictionary (like pip)
            if isinstance(dep, dict):
                continue

            # Split the dependency into name and version
            parts = dep.split('=')

            # Write the name and version to the requirements file
            f.write(f'{parts[0]}=={parts[1]}\n')

convert_conda_to_pip('/Users/lukasalemu/Documents/00. Bank of England/00. Degree/revisions_toolkit/revisions-toolkit/environment2.yml', '/Users/lukasalemu/Documents/00. Bank of England/00. Degree/revisions_toolkit/revisions-toolkit/requirements2.txt')
```
